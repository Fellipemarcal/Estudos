```python
stats_country = (
    df.groupby('country')
      .agg(
          users=('id', 'count'),
          total_paid=('price_paid', 'sum')
      )
      .sort_values('users', ascending=False)
      .head(10)
)

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.bar(stats_country.index, stats_country['users'])
ax1.set_ylabel('Nombre d’utilisateurs')
ax1.set_xlabel('Pays')

ax2 = ax1.twinx()
ax2.plot(stats_country.index, stats_country['total_paid'], marker='o')
ax2.set_ylabel('Montant total payé')

plt.title('Comparaison utilisateurs vs montant total par pays')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
