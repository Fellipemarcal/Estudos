```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = 'data.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.isnull().sum()
```

```python
df['tax'] = df['tax'].mode()[0] # on utilise remplir les valeurs null avec le valeur plus courant de la colonne 'tax'
df['country'] = df['country'].fillna('Inconnu') # ici on crée une nouvelle catégoire pour attribuer aux valeurs null
df['gender'] = df['gender'].fillna('Inconnu') # c'est la même chose ici
```

```python
df
```

```python
gender = (
    df
    .groupby('gender')
    .size()
    .sort_values(ascending = False)
)

gender
```

```python
country = (
    df
    .groupby('country')
    .size()
    .sort_values(ascending = False)
)

country
```

```python
plt.figure(figsize = (10,5))
gender.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.title('graphique gender')
plt.ylabel('quantité')
```

```python
plt.figure(figsize = (10,5))
gender.plot(kind='pie', autopct='%1.1f%%')
plt.xticks(rotation = 0)
plt.title('graphique gender')
plt.ylabel('quantité')
```
