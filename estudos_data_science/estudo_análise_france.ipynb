```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df['price_paid'] = (
    df['price_paid']
    .str.replace('$', '', regex = False)
    .astype(float)
)
```

```python
df['tax'] = df['tax'].mode()[0]
```

```python
df
```

```python
df['gender'] = df['gender'].fillna('Inconnu')
```

```python
df
```

```python
df['country'] = df['country'].fillna('Inconnu')
```

```python
df
```

```python
df.isnull().sum()
```

```python
france = df[df['country'] == 'France']
```

```python
france_gender = (
    france
    .groupby('gender')
    .size()
    .sort_values(ascending = False)
)
```

```python
france_gender
```

```python
france_male = france[france['gender'] == 'Male']
```

```python
france_male
```

```python
fr_gn_pp = (
    france
    .groupby('gender')['price_paid'].sum()
)
```

```python
fr_gn_pp
```

```python
plt.figure(figsize = (10,5))
fr_gn_pp.plot(kind = 'pie', autopct = '%1.1f%%')
plt.ylabel('')
plt.title('Contribuition par genre en France')
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y') # code pour pouvoir mettre la colonne date en ordre chrologique.
```

```python
fr_date = (
    france
    .groupby('date')
    .size()
    .sort_index(ascending = True) # dans cette partie il y a un petit difference que nous permet aussi mettre en ordre chronologique : sort_index()
                                  # à la place de sort_values()
)
```

```python
fr_date
plt.figure(figsize = (10,5))
fr_date.plot(kind = 'bar')
plt.title('Ventes par jour')
```

```python
plt.figure(figsize=(10,5))
fr_date.plot(kind='bar')

# mettre en évidence le max
bars = plt.gca().patches
max_bar = max(bars, key=lambda b: b.get_height())
max_bar.set_hatch('//')

plt.title('Ventes par date (pic mis en évidence)')
plt.xlabel('Date')
plt.ylabel('Montant total payé')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = fr_date.plot(kind='bar')

# récupérer les 5 plus grandes ventes
top5_values = fr_date.sort_values(ascending=False).head(5).values

# parcourir les barres et marquer celles du top 5
for bar in ax.patches:
    if bar.get_height() in top5_values:
        bar.set_hatch('//')

plt.title('Top 5 des meilleurs jours de ventes')
plt.xlabel('Date')
plt.ylabel('Montant total payé')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
